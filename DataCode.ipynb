```julia
using PyPlot
using JLD2
```

```julia
n = 300 ;
t = zeros(n) ;
for ii = 1 : n
    t[ii] = sqrt(ii) ;
end

xx = zeros(n) ;
yy = zeros(n) ;

for ii = 1 : n
    xx[ii] = t[ii]*cos(2*t[ii]) ;
    yy[ii] = t[ii]*sin(2*t[ii]) ;
end

data1 = zeros(n,2)
for ii = 1 : n
    data1[ii,1] = xx[ii] ;
    data1[ii,2] = yy[ii] ;
end

scatter(data1[:,1] , data1[:,2] , s = 2)
```

```julia
n = 100 ;
R = 5 ;
t = zeros(n) ;
for ii = 1 : n
    t[ii] = ii ;
end

t .*= 2*pi/n ;

X1 = zeros(n,3) ;
for ii = 1 : n
    X1[ii,1] = R*cos(t[ii]) + rand() ;
    X1[ii,2] = R*sin(t[ii]) + rand() ;
end

X2 = zeros(n,3) ;
for ii = 1 : n
    X2[ii,1] = R*cos(t[ii]) + rand() ;
    X2[ii,3] = R*sin(t[ii]) + rand() ;
end
X2[:,1] .-= 3*R/2 ;

X3 = zeros(n,3) ;
for ii = 1 : n
    X3[ii,1] = R*cos(t[ii]) + rand() ;
    X3[ii,3] = R*sin(t[ii]) + rand() ;
end
X3[:,1] .+= 3*R/2 ;

data2 = zeros(3*n,3) ;
data2[1:n,:] = X1 ;
data2[n+1:2*n,:] = X2 ;
data2[2*n+1:3*n,:] = X3 ;

scatter3D(data2[:,1],data2[:,2],data2[:,3],s = 20)
```

```julia
n = 30 ;

origin = [1 1 ; 1 -1 ; -1 1 ; -1 -1] ;
data3 = zeros(4*n,2) ;

for ii = 0 : 3
    for jj = 1 : n
        data3[ii*n+jj,1] = origin[ii+1,1] + rand() ;
        data3[ii*n+jj,2] = origin[ii+1,2] + rand() ;
    end
end

PyPlot.axis([-2,3,-2,3]) ;
scatter(data3[:,1],data3[:,2] , s = 5)
```
