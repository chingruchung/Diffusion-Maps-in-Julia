```julia
n = 300 ;
t = zeros(n) ;
for ii = 1 : n
    t[ii] = sqrt(ii) ;
end

xx = zeros(n) ;
yy = zeros(n) ;

for ii = 1 : n
    xx[ii] = t[ii]*cos(2*t[ii]) ;
    yy[ii] = t[ii]*sin(2*t[ii]) ;
end
```

```julia
using PyPlot
scatter(xx , yy , s = 2)
```

```julia
using JLD2
data = zeros(n,2)
for ii = 1 : n
    data[ii,1] = xx[ii] ;
    data[ii,2] = yy[ii] ;
end

@save "test.jld2"
```

```julia
n = 100 ;
R = 5 ;
t = zeros(n) ;
for ii = 1 : n
    t[ii] = ii ;
end

t .*= 2*pi/n ;

X1 = zeros(n,3) ;

for ii = 1 : n
    X1[ii,1] = R*cos(t[ii]) + rand() ;
    X1[ii,2] = R*sin(t[ii]) + rand() ;
end

X2 = zeros(n,3) ;

for ii = 1 : n
    X2[ii,1] = R*cos(t[ii]) + rand() ;
    X2[ii,3] = R*sin(t[ii]) + rand() ;
end

X2[:,1] .-= R ;
```

```julia
using PyPlot
scatter3D(X1[:,1],X1[:,2],X1[:,3])
scatter3D(X2[:,1],X2[:,2],X2[:,3])
```

```julia
using JLD2
data1 = zeros(2*n,3)

data1[1:n,:] = X1 ;
data1[1:n,:] = X1 ;

data1[n+1:2*n,:] = X2 ;
data1[n+1:2*n,:] = X2 ;

@save "test1.jld2"
```
